```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
%matplotlib notebook

plt.style.use('dark_background')
```

```python
train_df = pd.read_csv('data/TrainDF.csv').drop(['Item'], axis = 1)
test_df = pd.read_csv('data/TestDF.csv').drop('Item', axis = 1)

concat_df = train_df.append(test_df, ignore_index = True)
```

```python
# targets = concat_df.target.values.ravel()
features = concat_df[[i for i in concat_df.columns if i not in ['target', 'Cluster']]].values
# clusters = concat_df.Cluster.values.ravel()
pca = PCA(n_components = 3)
pca.fit(features)
# result = pca.fit_transform(features)

pos_df = concat_df[concat_df.target == 1]
neg_df = concat_df[concat_df.target == 0]

pos_targets = pos_df.target.values.ravel()
neg_targets = neg_df.target.values.ravel() 

pos_feat = pos_df[[i for i in pos_df.columns if i not in ['target', 'Cluster']]].values
neg_feat= neg_df[[i for i in neg_df.columns if i not in ['target', 'Cluster']]].values

pos_result = pca.transform(pos_feat)
neg_result = pca.transform(neg_feat)

pos_clusters = pos_df.Cluster.values.ravel()
neg_clusters = neg_df.Cluster.values.ravel()

colors = ['Orange', 'Yellow', 'Blue', 'Purple', 'Cyan', 'Pink']
pos_c = [colors[int(i)] for i in pos_clusters]
neg_c = [colors[int(i)] for i in neg_clusters]

markers = ['X', 's']
pos_m = [markers[int(i)] for i in pos_targets]
neg_m = [markers[int(i)] for i in neg_targets]

fig = plt.figure(figsize = (10,15))

ax = fig.add_subplot(111, projection = '3d')

ax.scatter(pos_result[:,0], pos_result[:,1], pos_result[:,2], c = pos_c, marker = 's', label = 'Positive')
ax.scatter(neg_result[:,0], neg_result[:,1], neg_result[:,2], c = neg_c, marker = '^', label = 'Negative')
ax.set_title('KMeans & PCA Clustering')
ax.legend()
leg = ax.get_legend()
leg.legendHandles[0].set_color('Green')
leg.legendHandles[1].set_color('Red')
```
